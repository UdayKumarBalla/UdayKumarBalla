## User Defined Function Section

The notebook can contain user defined functions. All user defined functions should be defined in the section below. There cannot be a main function in this section.

```python
# Functions Definition 
def sendemail():
    eam.log("in sendemail method start")
    msg = EmailMessage()
    msg['Subject'] = email_subject
    msg['From'] = 'noreply@hxgnsmartcloud.com'
    msg['To'] = email_to_list
    msg.add_alternative(email_body, subtype='html')
    # Connect to the email server
    with smtplib.SMTP(smtpserver) as server:
        server.send_message(msg)
    eam.log("in sendemail method end")
```

## Python Application Entry Point



This section should have only one function named main. This is the entry point for the application. This function can be edited to add user defined code and that code can call the user defined functions above. This function cannot be renamed and the parameter cannot be changed.


```python
def main(request=None):
    # User code starts here. Note that the print example below will display in the console 
    # when run in Python Studio but will not be accessible when run in batch execution.
    eam.log("Execution Starts!!!")
    sendemail()
    eam.log("Execution Ends!!!")
    
if __name__ == "__main__":
    #execute the script
    main()
```
